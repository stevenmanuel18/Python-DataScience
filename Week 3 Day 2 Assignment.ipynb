Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
# code here
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
fig= plt.figure(figsize=(10,8))
ax.plot(x, cos, color='r', label='Cosx')
ax.plot(x, cos3, color='g', ls='--', label='3 Cosx')
ax.plot(x, sin, color='b', ls='-.', label='Sin(90-x)')
ax.plot(x, sin2, color='y', ls=' ', marker='x', label='2 Sinx')
ax.legend(loc=10)
ax.set_xlabel('X-Axis Label')
ax.set_ylabel('Y-Axis Label')
ax.set_title('Grafik Cos Sin')
```
