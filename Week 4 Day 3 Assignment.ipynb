Buat model Klasifikasi (Model/Algoritma Bebas) untuk memprediksi status pasien dengan ketentuan sebagai berikut:

Baca https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_val_score.html untuk menggunakan metric roc_auc saat cross-validation.

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/haberman.csv'
list_cols = ['Age', "Patient's Years", "N_positive_ax", "survival_status"]
df = pd.read_csv(url, names=list_cols)
```

```python
df.head()
```

```python
df['survival_status'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

x = df.drop('survival_status', axis=1)
y = df['survival_status']
knn = KNeighborsClassifier(n_neighbors=3, weights='distance')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, stratify=y, random_state=64)
knn.fit(x_train, y_train)
y_predict = knn.predict(x_test)
```

```python
from sklearn.model_selection import cross_val_score

cv_results = cross_val_score(knn, x, y, cv=3)
cv_results.mean()
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, y_predict)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
confusion_matrix(y_test, y_predict)
```

```python
print(classification_report(y_test, y_predict))
```

1. Seberapa baik model anda dalam memprediksi seorang pasien mempunyai status positive?
    - Model memprediksi dengan akurasi 82%
2. Seberapa baik model anda dalam memprediksi seorang pasien mempunyai status negatif?
    - Model memprediksi dengan akurasi 43%

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/haberman.csv'
list_cols = ['Age', "Patient's Years", "N_positive_ax", "survival_status"]
df = pd.read_csv(url, names=list_cols)
```

```python
df.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
import numpy as np

X = df.drop('survival_status', axis=1)
y = df['survival_status']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=21)

model = KNeighborsClassifier()
```

```python
param_grid = {'n_neighbors':np.arange(3,50), 'weights':['distance', 'uniform'],
              'algorithm':['auto', 'ball_tree', 'kd_tree', 'brute'], 'leaf_size':[20,30,40]}
gscv = GridSearchCV(model, param_grid=param_grid, scoring='accuracy', cv=5)
gscv.fit(X,y)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
knn = KNeighborsClassifier(n_neighbors=37, weights='uniform')
_x = knn.fit(X_train, y_train)
_x
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, y_pred)
```

```python
_x.predict_proba(X_test)
```

```python
_x.classes_
```

```python
from sklearn.model_selection import cross_validate
score = cross_validate(knn, X, y, cv=5, return_train_score=True)
```

```python
score['test_score'].mean()
```

```python
score['train_score'].mean()
```

Model saya baik karena train score dengan accuracy 76.63%% dan test score dengan accuracy 77.13%

## DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV
```

```python
dtc = DecisionTreeClassifier()

param_rs = {'random_state':np.arange(0,60), 'criterion':['gini', 'entropy'], 'splitter':['best', 'random']}
rscv= RandomizedSearchCV(dtc,param_rs, n_iter=60 , scoring='accuracy', cv=5)
rscv.fit(X,y)
```

```python
rscv.best_score_
```

```python
rscv.best_params_
```
