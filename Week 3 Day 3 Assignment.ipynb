```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
fig, ax = plt.subplots(figsize=(20,12),ncols=3, sharex=True, sharey=True)
ax[0].plot(x, tan, color='b', label='tanX' )
ax[0].set_title('Grafik tanx')
ax[1].plot(x, sin, color='r', label='sinX')
ax[1].set_title('Grafik sinx')
ax[2].plot(x, cos, color='g', label='cosX')
ax[2].set_title('Grafik cosx')
fig.legend(loc=9)
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
fig, ax = plt.subplots(figsize=(16,12))
ax.plot(x,y)
ax.set_title('Grafik cosX')

top_right = ax.inset_axes([0.9,0.9,0.3,0.25])
top_right.plot(x, y4, color='k')
top_right.plot(x, y, color='b')
top_right.set_title('Grafik cosX^4')

bot_right = ax.inset_axes([0.9, -0.1, 0.3, 0.25])
bot_right.plot(x, y3, color='g')
bot_right.plot(x,y, color='b')
bot_right.set_title('Grafik cosX^3')

top_left = ax.inset_axes([-0.15, 0.9, 0.3, 0.25])
top_left.plot(x, y5, color='y')
top_left.plot(x,y, color='b')
top_left.set_title('Grafik cosX^5')

bot_left = ax.inset_axes([-0.1, -0.1, 0.3, 0.25])
bot_left.plot(x, y2, color='r')
bot_left.plot(x,y, color='b')
top_left.set_title('Grafik cosX^2')
```
